```python
def dotProduct(mtx1, mtx2):
   return mtx1.dot(mtx2)
```

```python
def cov(matrix):
    matrix_full_ones = np.full((matrix.shape[0], 1), 1)
    y_hat = (dotProduct(matrix.T, matrix) -(1/matrix.shape[0])*( dotProduct(matrix.T,matrix_full_ones)).dot(dotProduct(matrix.T, matrix_full_ones).T))
    result = (1/(matrix.shape[0]-1))
    return (result * y_hat)
```

```python

def corr(matrix):
    Dinv = np.diag(1 / np.sqrt(np.diag(matrix)))
    corr = Dinv @ matrix @ Dinv
    return corr
```

```python
corr(cov(data))
```

*Variance:* 

The variance is the measure of how far the individual values in a data are from the mean value. 

Is used to quantify spread or dispersion that describes how much variability there is in it.

If we have a high variance, this tells us that the values in our data are far from their mean, having a high variability.

If we have a low variance, this tells us that the values are close to the mean, having a low variability.

Covariance:

The covariance is a measure of the strength of the correlation between two or more sets of random variates. It evaluates how much to what extent the variables change together. 

If we have a positive covariance between 2 variables this means that this 2 variables tend to move in the same direction.

If we have a negative covariance between 2 variables this means that this 2 variables tend to move in inverse directions.

Correlation:

The correlation is a measure that expresses the extent to which two variables are linearly related, this means they change together at a constant rate.

It is measured by the coefficient "r" that is a value between -1 and +1 that quantifies the strength of the relationship between 2 variables.

The closer the value is to 1 or -1, the stronger the relationship. This means that as one value increases, so does the other one.

The closer the value is to 0, the weaker the relationship. This means that as one value increases, the other decreases.

```python
def getL(l):
    sum = l.sum()
    shape = l.shape[0]
    return sum/shape
```

```python
def leveragePoints(mtx):
    leverages = np.diagonal(mtx.dot(np.linalg.inv(mtx.T .dot(mtx)).dot(mtx.T)))
    boolean_mtx = leverages > (3 * getL(leverages))
    real_values_of_mtx = np.nonzero(boolean_mtx)
    return real_values_of_mtx[0]
```

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn import preprocessing

def calculate_vif(df, features):    
    df = df.dropna()
    vif, tolerance = {}, {} 
    for feature in features:
        
        X = [f for f in features if f != feature]        
        X, y = df[X], df[feature]  
        r2 = LinearRegression().fit(X, y).score(X, y)                
        
        
        tolerance[feature] = 1 - r2
        vif[feature] = 1/(tolerance[feature])
    return pd.DataFrame({'VIF': vif, 'Tolerance': tolerance})
```

Multicollinearity analysis:

Multicollinearity is the strong linear dependence relationship between more than 2 groups of explanatory variables in a multiple regression.

In other words, it is the high correlation between two or more explanatory variables.

The greater the value is closer to 1, this means that the variables are highly correlated.

```python
X = data[['booktomarketratio','salesannualgrowth', 'operatingprofitgrowth']]
calculate_vif(df=X, features=['booktomarketratio','salesannualgrowth', 'operatingprofitgrowth'])
```
